```python
mpg = sbn.load_dataset('mpg')
mpg
```

```python
sbn.catplot(x = 'cylinders', y = 'horsepower', data = mpg, hue = 'origin', palette = 'OrRd', height = 6, aspect = 6/4)
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (15,7))
sbn.set_style('whitegrid')
sbn.barplot(x = 'cylinders', y = 'horsepower', data = mpg, hue = 'origin', palette = 'OrRd', )
# plt.grid()
plt.show()
```

```python
plt.figure(figsize = (20,5))
sbn.set_style('whitegrid')
sbn.set_context("poster", font_scale = 1)
sbn.barplot(x = 'cylinders', y = 'horsepower', data = mpg, hue = 'origin', palette = 'OrRd', )
sbn.despine()
plt.show()
```

```python
plt.figure(figsize = (10,5))
sbn.set_style('whitegrid')
sbn.set_context("paper", font_scale = 1.5)
sbn.barplot(x = 'cylinders', y = 'horsepower', data = mpg, hue = 'origin', palette = 'OrRd', )
sbn.despine()
plt.show()
```

```python
fg = sbn.FacetGrid(mpg, col = 'origin', hue = 'cylinders', palette = 'autumn')
fg.map(plt.bar, 'cylinders', 'horsepower', edgecolor = 'r').add_legend()
sbn.set_style('whitegrid')
sbn.set_context("paper", font_scale = 1)
sbn.despine()
plt.show()
```
