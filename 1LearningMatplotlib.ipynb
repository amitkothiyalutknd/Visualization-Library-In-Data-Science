```python
from numpy import *
def f(t):
    return exp (-t) * cos(2* pi *t)
t1 = arange(0.0, 5.0, 0.1)
t2 = arange(0.0, 5.0, 0.02)
plt.subplot(211)
plt.plot(t1,f(t1), 'bo', t2, f(t2))
plt.subplot(212)
plt.plot(t2, cos(2*pi*t2))
plt.show()
```
